```python
import os
import re

# Specify the directory containing the files
folder_path = r"C:\Users\liang\Documents\GitHub\fusion\experiments_synthetic\results\regression_quadratic_early"

# Loop through each file in the directory
for filename in os.listdir(folder_path):
    # Match the original filenames like 'noisy_dim200_seed1.txt'
    match = re.match(r'noisy_dim(\d+)_seed(\d+)\.txt', filename)
    
    if match:
        dim_num = match.group(1)  # Extract the dimension number
        seed_num = match.group(2)  # Extract the seed number
        
        # Construct the new filename with the added pattern
        new_filename = f"noisy_dim{dim_num}_noise_ratio0.5_scale0_seed{seed_num}.txt"
        
        # Get full paths for renaming
        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_filename)
        
        # Rename the file
        os.rename(old_file, new_file)
        print(f"Renamed: {filename} -> {new_filename}")
```
